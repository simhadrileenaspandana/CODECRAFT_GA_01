```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Importing libraries needed for GPT-2 text generation

# transformers:
# This is a Python library developed by Hugging Face.
# It provides many pre-trained AI models like GPT-2, BERT, T5 etc.
# We use it because it allows us to easily load and use powerful AI models.


# GPT2LMHeadModel:
# GPT-2 is a language model used for text generation.
# "LM" stands for Language Model.
# "Head" refers to the final neural network layer that predicts the next word.
# So GPT2LMHeadModel = GPT-2 model + language prediction layer.
# We use this class when we want GPT-2 to generate text.


# GPT2Tokenizer:
# Computers do not understand raw text.
# A tokenizer converts text into numbers (tokens) that the model understands.
# Example:
# "AI is powerful" -> [15496, 318, 3665]
# The tokenizer also converts tokens back into readable text.


# torch:
# torch is the main library of PyTorch.
# PyTorch is a deep learning framework used to run neural networks.
# GPT-2 internally uses PyTorch to perform mathematical computations
# and to run the neural network model.

```

```python
# Loading the GPT-2 model and tokenizer

# from_pretrained("gpt2") means:
# We are downloading a pretrained GPT-2 model that was already trained
# on a very large dataset from the internet.

# This saves us from training a model from scratch.

model = GPT2LMHeadModel.from_pretrained("gpt2")

# Load the tokenizer for GPT-2
# The tokenizer converts text into tokens (numbers) that the model understands.

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Print confirmation so we know everything loaded correctly
print("GPT-2 model and tokenizer loaded successfully!")
```

```python
# Step 5: Give a prompt to GPT-2

# A prompt is the starting text we give to the model.
# The model will continue the sentence based on what it learned during training.

prompt = "Artificial Intelligence will"

# The tokenizer converts the prompt into tokens (numbers) that the model understands.
# return_tensors="pt" means we want the tokens in PyTorch format.

inputs = tokenizer.encode(prompt, return_tensors="pt")

# Let's print the tokens to see how the text is converted into numbers
print(inputs)
```

```python
# Step 6: Generate text using GPT-2

# The generate() function tells GPT-2 to continue the sentence
# based on the prompt we provided earlier.

outputs = model.generate(
    inputs,                # tokens from the prompt
    max_length=50,         # maximum length of generated text
    num_return_sequences=1 # number of outputs we want
)

# Convert tokens back into readable text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Print the generated text
print(generated_text)
```

```python
# Step 7: Generate multiple outputs with sampling

outputs = model.generate(
    inputs,
    max_length=50,
    num_return_sequences=3,   # generate 3 outputs
    do_sample=True            # enables sampling instead of greedy search
)

for i, output in enumerate(outputs):
    text = tokenizer.decode(output, skip_special_tokens=True)
    print(f"Output {i+1}:")
    print(text)
    print()
```

```python
# Step 8: Create a custom dataset file

dataset_text = """
Success comes from hard work and dedication.
Never stop learning new things in life.
Believe in yourself and your abilities.
Consistency and patience lead to success.
Dream big and work hard every day.
Great achievements start with small steps.
Hard work always pays off in the end.
Stay positive and keep moving forward.
Failure is just another step toward success.
Confidence and discipline build strong leaders.
"""

with open("dataset.txt", "w") as file:
    file.write(dataset_text)

print("Dataset file created successfully!")
```

```python
# Step 9: Load and read the dataset

# Open the dataset file we created
with open("dataset.txt", "r") as file:
    data = file.read()

# Print the dataset content
print("Dataset content:\n")
print(data)
```

```python
# Step 10: Convert dataset text into tokens

# Tokenize the dataset text so the model can understand it
dataset_tokens = tokenizer.encode(data, return_tensors="pt")

# Print the tokens
print(dataset_tokens)
```

```python
# Step 11: Fine-tuning GPT-2 with our dataset

# Set the model in training mode
model.train()

# Define optimizer (used to update the model weights during training)
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# We will train for a few steps since our dataset is small
for epoch in range(3):  # training for 3 rounds
    outputs = model(dataset_tokens, labels=dataset_tokens)

    loss = outputs.loss
    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

    print(f"Epoch {epoch+1} Loss:", loss.item())
```

```python
# Step 12: Generate text from the fine-tuned model

# Give a prompt related to our dataset
prompt = "Success in life"

# Convert the prompt into tokens
inputs = tokenizer.encode(prompt, return_tensors="pt")

# Generate text using the fine-tuned model
outputs = model.generate(
    inputs,
    max_length=50,
    num_return_sequences=3,
    do_sample=True
)

# Convert tokens back to readable text
for i, output in enumerate(outputs):
    text = tokenizer.decode(output, skip_special_tokens=True)
    print(f"Generated Text {i+1}:")
    print(text)
    print()
```
